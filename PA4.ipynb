### Problem #1

```python
# import library
import pandas as pd

# load the dataset
df = pd.read_excel("board2.xlsx")

# a) filter: Instrumentation, Luzon, Electronics > 70
res_a = df.loc[
    (df["Track"] == "Instrumentation") &
    (df["Hometown"] == "Luzon") &
    (df["Electronics"] > 70),
    ["Name", "GEAS", "Electronics"]
]
display(res_a)

# b) compute row average for selected subjects
df["Avg"] = df[["Math", "Electronics", "GEAS", "Communication"]].mean(axis=1)

# filter: Mindanao, Female, Avg >= 55
res_b = df.loc[
    (df["Hometown"] == "Mindanao") &
    (df["Gender"] == "Female") &
    (df["Avg"] >= 55),
    ["Name", "Track", "Electronics", "Avg"]
]
display(res_b)
```

### Problem #2

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Track vs Average
plt.figure(figsize=(8,7))
sns.barplot(x="Track", y="Avg", data=df, estimator="mean")
plt.title("Average Score by Track")
plt.ylabel("Average Scores")
plt.show()
```

```python
# Gender vs Average
plt.figure(figsize=(8,7))
sns.barplot(x="Gender", y="Avg", data=df, estimator="mean")
plt.title("Average Score by Gender")
plt.ylabel("Average Scores")
plt.show()
```

```python
# Hometown vs Average
plt.figure(figsize=(8,7))
sns.barplot(x="Hometown", y="Avg", data=df, estimator="mean")
plt.title("Average Score by Hometown")
plt.ylabel("Average Scores")
plt.show()
```
